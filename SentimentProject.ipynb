```python
# Define columns to store sentiment results
sentiment_columns = ['sentiment', 'compound_score', 'positive_score']

import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Loop through each text entry and analyze sentiment
for index, row in df.iterrows():
    text = row['CleanText']
    sentiment = analyzer.polarity_scores(text)

    # Extract relevant scores and store them in corresponding columns
    df.loc[index, sentiment_columns] = sentiment['compound'], sentiment['pos'], sentiment['pos']

print(df)
```

```python
# Calculate sentiment distribution with error handling
positive_count = len(df[df['Sentiment'] == ' Positive  '])
negative_count = len(df[df['Sentiment'] == ' Negative  '])
neutral_count = len(df[df['Sentiment'] == ' Neutral   '])
```

```python
# Small value to avoid division by zero (optional)
epsilon = 0.001  # Modify if needed
```

```python
# Check for empty DataFrame
if df.empty:
    print("Error: DataFrame is empty. No sentiment analysis possible.")
else:
    total_count = positive_count + negative_count + neutral_count + epsilon
```

```python
# Calculate percentages with safe division
positive_pct = (positive_count / total_count) * 100
negative_pct = (negative_count / total_count) * 100
neutral_pct = (neutral_count / total_count) * 100

print(f"Positive: {positive_pct:.2f}%, Negative: {negative_pct:.2f}%, Neutral: {neutral_pct:.2f}%")
```

```python
# Visualizations:
import matplotlib.pyplot as plt

# Bar Chart
sentiment_counts = [positive_count, negative_count, neutral_count]
sentiment_labels = ['Positive', 'Negative', 'Neutral']

plt.bar(sentiment_labels, sentiment_counts)
plt.xlabel("Sentiment")
plt.ylabel("Number of Texts")
plt.title("Sentiment Distribution")
plt.show()
```

```python
# Pie Chart
plt.pie(sentiment_counts, labels=sentiment_labels, autopct="%1.1f%%")
plt.title("Sentiment Distribution")
plt.show()
```

```python
# Histogram of Compound Sentiment Scores
plt.hist(df['compound_score'], bins=20)
plt.xlabel("Compound Sentiment Score")
plt.ylabel("Frequency")
plt.title("Distribution of Sentiment Scores")
plt.show()
```

```python
# Box Plots of Sentiment Scores by Sentiment Category:
plt.boxplot([df[df['Sentiment'] == ' Positive  ']['compound_score'],
             df[df['Sentiment'] == ' Negative  ']['compound_score'],
             df[df['Sentiment'] == ' Neutral   ']['compound_score']],
            labels=['Positive', 'Negative', 'Neutral'])
plt.title("Compound Sentiment Scores by Sentiment Category")
plt.show()
```

```python
#  Scatter Plot of Compound Score vs. Positive Score
plt.scatter(df['compound_score'], df['positive_score'])
plt.xlabel("Compound Sentiment Score")
plt.ylabel("Positive Sentiment Score")
plt.title("Relationship Between Compound and Positive Sentiment")
plt.show()
```

```python
# Word Cloud of Most Frequent Words in Each Sentiment Category
from wordcloud import WordCloud

for sentiment in [' Positive  ', ' Negative  ', ' Neutral   ']:
    words = ' '.join(df[df['Sentiment'] == sentiment]['CleanText'])
    wordcloud = WordCloud(width=800, height=400).generate(words)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud for {sentiment} Sentiment")
    plt.show()
```
